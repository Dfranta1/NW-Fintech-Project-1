```python
# Create variable and pull in data for financial ratios
ratios_df = sf.load_derived(variant='quarterly-asreported', market='us')
print(ratios_df.head())
```

```python
# Drop unneeded columns
    # SimFinID has no relation to stock performance
    # All stock ratios are in US dollars
    # Restated date
    # Dividends Per Share and Dividends Payout Ratio filled with NaN or 0.0
ratios_df.drop(columns=['SimFinId', 'Currency', 'Restated Date', 'Dividends Per Share', 'Dividend Payout Ratio'], axis=1, inplace=True)
ratios_df
```

```python
# Create variable and pull in data for share prices ratios
prices_df = sf.load_shareprices(variant='daily', market='us')
```

```python
prices_df.info()
```

```python
# Reset index for share price dataframe
prices_df.reset_index(inplace=True)
prices_df
```

```python
# Format the date time
prices_df.index = pd.to_datetime(prices_df.Date)
```

```python
# Groupby Ticker so percent change in share price is for each individual stock
prices_df = prices_df.groupby("Ticker").resample("Q").last()
prices_df
```

```python
# Check stock prices dataframe information
prices_df.info()
```

```python
#Calculate percentage in stock price before merging
prices_df['returns'] = prices_df["Adj. Close"].groupby('Ticker').pct_change().shift(-1)
prices_df
```

```python
# Join the dataframes through intersectinon
intersect = ratios_df.index.intersection(prices_df.index)
```

```python
# Prep dataframes for concatenation
ratios_df = ratios_df.loc[intersect, :].sort_index()
prices_df = prices_df.loc[intersect, :].sort_index()
```

```python
# Create new dataframe for the combined stock price and ratio dataframes
concat_df = ratios_df.join(prices_df)
```

```python
# View new dataframe information
concat_df.info()
```

```python
# Drop NA's
concat_df.dropna(inplace=True)
concat_df
```
